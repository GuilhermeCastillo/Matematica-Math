<h4>Um conjunto é formado por elementos<h4>

<p>
    A é um conjunto qualquer<br/>
    a --> é um objeto qualquer

<h6>Existe uma relação entre eles:</h6>

    
$ a \in A $ : o elemento a pertence ao conjunto A
<br/>
$ a \not\in A : $ o elemento a não pertence ao conjunto A


```python
# conjunto
A = set([1, 2, 3, 4])
# elemento
a = 2

if a in A:
    print(
    '''
    o elemento 2 está contido em no conjunto A.
    ''')
```

```python
# conjunto
A = set([1, 2, 3, 4])
# elemento
a = 100

if a in A:
    pass
else:
    print(
    '''
    o elemento 100 não está contido em no conjunto A.
    ''')
```

<h1>Relação de Inclusão</h1>

<p>
    Sejam A e B conjuntos. Se todo elemento de A for elemento também de B, diz-se que A é um subconjunto de B, que A está contido em B.
</p>

$ A \subset B $ : o conjunto A está contido em B

$ A \not\subset B $ :  Nem todo elemento de A pertence a B

```python
A = set([1, 2, 3, 4, 100, 200, 300, 400])
B = set([1, 2, 3, 4, 100, 200, 300, 400,100,200,300])

contador_elem = 0
for elem in A:
    if elem in B:
        contador_elem += 1

if contador_elem == len(A):
    print("A está contido em B")
```

```python
A = set([1, 2, 3, 4, 100, 200, 300, 400,4000,6000])
B = set([1, 2, 3, 4, 100, 200, 300, 400,100,200,300])

contador_elem = 0
for elem in A:
    if elem in B:
        contador_elem += 1

if contador_elem != len(A):
    print("Nem todo elemento de A pertence a B.")
```

<h6>Para todo conjunto A, vale:</h6>

$ \emptyset \subset A$ 

<h6>Propriedades fundamentais da Relação de Inclusão:</h6>

<p>Dados os conjuntos: A, B, C</p>

$ A \subset A $ : Reflexividade

Se $A \subset B$  e  $ B \subset A $  então A = B : Anti-Simetria

Se $ A \subset B $ e $ B \subset C $ então $ A \subset C $ : Transitividade

<h1>Reunião e Interseção</h1>

<p>
    Dados os conjuntos A e B,  a reunião é o conjunto formado por todos os elementos de A mais os elementos de B.
</p>

$ A \cup B$ 

```python
A = set([1, 2, 3, 4, 100, 200, 300, 400,4000,6000])
B = set([1, 2, 3, 4, 100, 200, 300, 400,100,200,300])

A.union(B)
```

<p>Enquanto que a interseção é o conjunto de objetos que são ao mesmo tempo de A e de B.</p>

$A \cap B$

```python
A.intersection(B)
```

x $ \in A \cup B $ : " x $ \in A $ ou x $ \in B $"

x $ \in A \cap B $ : " x $ \in A $ e  x $ \in B $"
