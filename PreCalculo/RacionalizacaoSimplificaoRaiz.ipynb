```python
from functools import reduce
```

```python
def fatoracao_de_um_numero(numero):
  fatoracao = []

  divisor = 2
  fatoracao = []
  while True:
    if numero % divisor == 0:
      numero = numero / divisor
      fatoracao.append(divisor)
    else:
      divisor += 1
    
    if numero == 1:
      break
  
  return fatoracao
```
