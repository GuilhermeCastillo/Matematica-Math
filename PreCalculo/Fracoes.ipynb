<h1>Frações</h1>


```python
class Fracao:

  def __init__(self, fracao_A, fracao_B):
    self.numeradorA = fracao_A[0]
    self.denominadorA = fracao_A[1]
    self.numeradorB = fracao_B[0]
    self.denominadorB = fracao_B[1]

  # Operações com frações
  def soma(self):
    resultado = None
    # Denominadores iguais
    if self.denominadorA == self.denominadorB:
      soma_numerador = self.numeradorA + self.numeradorB
      resultado = round(soma_numerador / self.denominadorA, 2)
    # Denominadores Diferentes
    else:
      formula_resultado = (self.numeradorA * self.denominadorB + self.numeradorB * self.denominadorA) / (self.denominadorA * self.denominadorB)
      resultado = round(formula_resultado, 2)
    return resultado

  def subtracao(self):
    resultado = None
    # Denominadores iguais
    if self.denominadorA == self.denominadorB:
      soma_numerador = self.numeradorA - self.numeradorB
      resultado = round(soma_numerador / self.denominadorA, 2)
    # Denominadores Diferentes
    else:
      formula_resultado = (self.numeradorA * self.denominadorB + self.numeradorB - self.denominadorA) / (self.denominadorA * self.denominadorB)
      resultado = round(formula_resultado, 2)
    return resultado

  def multiplicacao(self):
    resultado = round((self.numeradorA * self.numeradorB) / (self.denominadorA * self.denominadorB), 2)
    return resultado
  
  def divisao(self):
    resultado = round((self.numeradorA * self.denominadorB) / (self.denominadorA * self.numeradorB), 2)
    return resultado
```

```python
a = Fracao([6,10], [5,9])

a.multiplicacao()
```

```python
b = Fracao([4,5], [3,5])

b.soma()
```

```python
c = Fracao([1,6], [5,6])

c.subtracao()
```

```python
d = Fracao([3,12], [9,4])

d.divisao()
```
